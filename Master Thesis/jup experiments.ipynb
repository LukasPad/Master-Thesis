```python


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


feats = pd.read_csv('impr-.csv')
feats2 = pd.read_csv('impr+.csv')
attrs = pd.read_csv('attrs.csv')
clos = pd.read_csv('clos.csv')
adapV = pd.read_csv('adapV.csv')
adapE = pd.read_csv('adapE.csv')
adapc = pd.read_csv('adap comp.csv')
```

```python
ordered_feats2 = feats2.sort_values('edges').reset_index(drop = True)
ordered_feats = feats.sort_values('edges').reset_index(drop = True)
attrs = attrs.sort_values('edges').reset_index(drop = True)
clos = clos.sort_values('edges').reset_index(drop = True)

ordered_feats = ordered_feats.loc[:, ~ordered_feats.columns.str.contains('^Unnamed')]
ordered_feats2 = ordered_feats2.loc[:, ~ordered_feats2.columns.str.contains('^Unnamed')]
```

```python
comparisonp = ordered_feats2.copy()
for x in range(0,comparisonp.shape[0]):
    comparisonp.iloc[x, 4:] = ((comparisonp.iloc[x, 4:] - comparisonp.iloc[x, 3])/comparisonp.iloc[x, 3]) * 100
display(comparisonp)
comparisonm = ordered_feats.copy()

for x in range(0,comparisonm.shape[0]):
    comparisonm.iloc[x, 4:] = ((comparisonm.iloc[x, 4:] - comparisonm.iloc[x, 3])/comparisonm.iloc[x, 3]) * 100
comparisonm
display(comparisonm.sort_values('all'))
```

```python
import seaborn as sns

sets = [attrs.iloc[:, 2:], comparisonp.iloc[:, 3:], comparisonm.iloc[:, 3:]]
full_df = pd.concat(sets, axis=1)
corr_df = full_df.corr()
sns.heatmap(corr_df)
```

```python
display(corr_df)
full_df
```

```python

dom_scat = plt.scatter(attrs['avg deg'], full_df['w/o dom rule'])

#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(attrs['avg deg'], full_df['w/o dom rule'], 1)
m22, m21, b2 = np.polyfit(attrs['avg deg'], full_df['w/o dom rule'], 2)
m33, m32 ,m31, b3 = np.polyfit(attrs['avg deg'], full_df['w/o dom rule'], 3)
m44, m43, m42 ,m41, b4 = np.polyfit(attrs['avg deg'], full_df['w/o dom rule'], 4)

x = np.linspace(1, 2.5, 100)
y1 = np.polyval([m, b], x)
y2 = np.polyval([m22, m21, b2], x)
y3 = np.polyval([m33, m32, m31, b3], x)
y4 = np.polyval([m44, m43, m42 ,m41, b4], x)

#add linear regression line to scatterplot 
plt.scatter(1.47531, 0, s=200, marker='x', zorder=10, color='black')
plt.plot(x, y1, color='g', label='degree-1')
plt.plot(x, y2, color='orange', label='degree-2')
plt.plot(x, y3, color='cyan', label='degree-3')
plt.plot(x, y4, color='purple', label='degree-4')
plt.axhline(y=0, color='r')
plt.text(1.27531, 30, 'intercept')
plt.legend()
```

```python
adapV = adapV.loc[:, ~adapV.columns.str.contains('^Unnamed')]

adapV2 = adapV.iloc[:, 1:]
adapV2 = adapV2.sort_values('1.0').reset_index(drop = True)
adapV2 = adapV2.div(adapV2.sum(axis=1), axis=0)


for i, col in enumerate(adapV2.columns):
    #adapV2[col] = (adapV2[col])/(adapV2[col].sum())
    plt.plot(adapV2.index,adapV2[col], label=col)

plt.legend()
plt.show()
adapV2
adapV3 = adapV.iloc[:, 1:].sum(axis=0)
adapV3
```
